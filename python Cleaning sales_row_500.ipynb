```python
#data load

import pandas as pd

df = pd.read_excel(r"D:\power bi desktop\Power bI project\inamul data analytics end to end project\sales_raw_500.xlsx")
```

```python
df
```

```python
#duplicated removes

df = df.drop_duplicates()
```

```python
df
```

```python
#fix date column

df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

```python
df = df.dropna(subset=['Order_Date'])
```

```python
df
```

```python
df['Profit'] = df['Sales']-df['Cost']
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

```python
df.to_csv('Sales_cleaned.csv',index = False)
```
